Part 1: parse rules.

Your goal is to determine the number of messages that completely match rule 0.

```python
input=open("input.txt")
input = input.readlines()
input = [x.strip() for x in input]
print(len(input))
print(len(input[1]))
print(input[:10])
```

```python
rules = {}
for x in range(len(input)):
    if input[x] == '':
        messages = input[x+1:]
        break
    rule_no, rules_to_match = input[x].split(':')
    if rules_to_match.strip()[0] == '"':
        rules[rule_no] = rules_to_match.strip()[1]
    else:
        rules[rule_no] = [rule_set.strip().split() for rule_set in rules_to_match.split('|')]
    #print(rule_no, rules)
```

```python
def all_valid(message, rule_list):
    match_pos = -1
    for rule in rule_list:
        if valid(message[match_pos+1:],rule) >= 0:
            match_pos += 1 + valid(message[match_pos+1:],rule)
        else:
            #this is a failed match
            #print(message, rule_list, match_pos)
            return -1
    #all passed so return end position of final match
    #print(message, rule_list, match_pos)
    return(match_pos)
```

```python
#function to check if a message matches a rule and return the matched digit count if so
def valid(message,rule_no):
    #print(message,rule_no,rules[rule_no])
    if message == '':
        return -1
    rules_to_match = rules[rule_no]
    if rules_to_match == 'a':
        if message[0] == 'a':
            return 0
        else:
            return -1
    elif rules_to_match == 'b':
        if message[0] == 'b':
            return 0
        else:
            return -1
    else:
        #check if any of the sets of rules listed are matched
        matches = [all_valid(message,rule_list) for rule_list in rules_to_match]
        #print(message, rules_to_match, 'matches', matches)
        if max(matches) >= 0:
            return min([match for match in matches if match >= 0])
        else:
            return -1
```

```python
match_count = 0
for message in messages:
    #print(message, valid(message,'0'))
    if valid(message,'0') == len(message) - 1:
        match_count += 1
```

Part 2: new rules; may imply loops...

8: 42 | 42 8

11: 42 31 | 42 11 31

```python
rules['8'] = [['42'],['42','8']]
rules['11'] = [['42','31'],['42','11','31']]
```

```python
#if all are valid, return list of locations where they can be
def all_valid(message, rule_list):
    if len(rule_list) == 1:
        return(valid(message,rule_list[0]))
    else:
        #2 rules only other option except one instance of 3 rules
        possibilities = {-1}
        #catch if more rules than characters
        if len(message) < len(rule_list):
            return([-1])
        first = valid(message,rule_list[0])
        for item in first:
            if (0 <= item < len(message) -1):
                #otherwise just skip to return the -1 since no ways can work
                second = valid(message[item+1:],rule_list[1])
                for second_item in second:
                    if second_item >= 0:
                        if len(rule_list) == 2:
                            possibilities.add(item + 1 + second_item)
                        else:
                            #length 3 case
                            third = valid(message[item+1+second_item+1:],rule_list[2])
                            for third_item in third:
                                if third_item >= 0:
                                    possibilities.add(item + 1 + second_item + 1 + third_item)
        return(list(possibilities))
            
    
```

```python
def valid(message,rule_no):
    #print(message,rule_no,rules[rule_no])
    if message == '':
        #print('test')
        return [-1]
    rules_to_match = rules[rule_no]
    if rules_to_match == 'a':
        if message[0] == 'a':
            return [0]
        else:
            return [-1]
    elif rules_to_match == 'b':
        if message[0] == 'b':
            return [0]
        else:
            return [-1]
    else:
        #check if any of the sets of rules listed are matched
        matches = []
        for rule_list in rules_to_match:
            matches.append(all_valid(message,rule_list))
        #print(message, rules_to_match, 'matches', matches)
        #return unique possibilities
        if len(matches) > 1:
            matches = list(set(matches[0]+matches[1]))
        else:
            matches = matches[0]
        return(matches)
    
```
