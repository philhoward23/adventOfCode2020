```python

#part 1 cid can be skipped for north pole passports
day6input=open("input.txt")
input = day6input.readlines()
input = [x.strip() for x in input]
```

```python
print(len(input))
input[:10]
```

```python
groups = [set()]
for line in input:
    if (line == ''):
        groups.append(set())
    else:
        groups[-1].update(a for a in line)
        
```

```python
print(len(groups))
groups[:3]
```

```python
[len(group) for group in groups[-5:]]
```

```python
sum([len(group) for group in groups])
```

Part 2: need sum of counts of ALL party members answering the same

```python
groups_all = [{}]
flag = True
for line in input:
    if (line == ''):
        flag = True
        groups_all.append({})
    elif flag:
        flag = False
        groups_all[-1] = set(a for a in line)
    else:
        groups_all[-1].intersection_update(a for a in line)
        
```

```python
groups_all[:10]
```

```python
sum([len(group) for group in groups_all if len(group)>=1])
```

```python
[len(group) for group in groups_all]
```
