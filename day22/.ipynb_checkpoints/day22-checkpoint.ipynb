```python
#play
while (len(player_one) > 0) and (len(player_two) > 0):
    one = player_one[0]
    two = player_two[0]
    if one > two:
        player_one = player_one[1:]+[one,two]
        player_two = player_two[1:]
    else:
        player_one = player_one[1:]
        player_two = player_two[1:]+[two,one]
    print(player_one, player_two)    
```

```python
def score(deck):
    score = 0
    for jj in range(len(deck)):
        score += deck[jj]*(len(deck) - jj)
    return(score)
```

```python
import collections
```

```python
import copy
def play_round(deck_one, deck_two,games):
    card_one = deck_one.popleft()
    card_two = deck_two.popleft()
    if ((card_one <= len(deck_one)) and (card_two <= len(deck_two))):
        if score(play_game(collections.deque(copy.deepcopy(deck_one)), collections.deque(copy.deepcopy(deck_two)),games)[0]) > 0:
            #one won
            deck_one.append(card_one)
            deck_one.append(card_two)
        else:
            deck_two.append(card_two)
            deck_two.append(card_one)
    else:
        if card_one > card_two:
            deck_one.append(card_one)
            deck_one.append(card_two)
        else:
            deck_two.append(card_two)
            deck_two.append(card_one)  
    return(deck_one,deck_two)
```

```python
import copy
def play_game(one,two,games):
    games[0] += 1
    if games[0] % 10 == 0:
        print(games, one, two)
    #configs = ''
    configs = set()
    while (len(one) > 0) and (len(two) > 0):
        #check if this is an original configuration so ok to play round
        #if (str(one) + str(two)) in configs:
        if (tuple(one),tuple(two)) in configs:
            #one wins
            print("Breaking loop", len(configs))
            two = []
        else:
            #configs+=('_'+str(one)+str(two))
            configs.add((tuple(one),tuple(two)))
            #one, two = play_round(one, two,games)
            card_one = one.pop(0)
            card_two = two.pop(0)
            if ((card_one <= len(one)) and (card_two <= len(two))):
                if score(play_game(one[:card_one], two[:card_two],games)[0]) > 0:
                    #one won
                    one.append(card_one)
                    one.append(card_two)
                else:
                    two.append(card_two)
                    two.append(card_one)
            else:
                if card_one > card_two:
                    one.append(card_one)
                    one.append(card_two)
                else:
                    two.append(card_two)
                    two.append(card_one)             
        #print(one, two)   
    #print(score(one), score(two))
    return(one,two)
```

I had missed that the recursive game only used decks of size equal to the cards drawn

```python
from collections import deque
from itertools import islice

rounds = [0]

with open("test_input_3.txt", "r") as f:
    first, second = map(str.splitlines, f.read().split('\n\n'))
    p1, p2 = deque(map(int, first[1:])), deque(map(int, second[1:]))

SCORE = lambda cards: sum(i*card for i, card in enumerate(reversed(cards), 1))


def part1(p1, p2):
    while p1 and p2:
        one, two = p1.popleft(), p2.popleft()
        p1.extend([one, two]) if one > two else p2.extend([two, one])
    
    return p1 or p2


def recursive_combat(p1, p2):
    rounds [0] +=1
    if rounds[0] % 10 == 0:
        print(rounds, p1, p2)
    seen = set()
    while p1 and p2:
        if (prev:= (tuple(p1), tuple(p2))) in seen: return True
        seen.add(prev)

        one, two = p1.popleft(), p2.popleft()
        if len(p1) >= one and len(p2) >= two:
            result = recursive_combat(deque(islice(p1, one)), deque(islice(p2, two)))
            p1.extend([one, two]) if result else p2.extend([two, one])
            continue
            
        p1.extend([one, two]) if one > two else p2.extend([two, one])
        #print(p1,p2)
    
    return bool(p1)


def part2(p1, p2):
    return p1 if recursive_combat(p1, p2) else p2


print(SCORE(part1(p1.copy(), p2.copy())))
print(SCORE(part2(p1.copy(), p2.copy())))
```
