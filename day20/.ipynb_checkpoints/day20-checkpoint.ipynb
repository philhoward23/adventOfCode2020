#12x12 grid needed

```python
class tile:
    def __init__(self, id, array):
        self.id = id
        self.array = array
        #sides: [top, bottom, right, left, fliptop, flipbottom, flipright, flipleft]
        self.sides = []
        self.sides.append(self.array[0])
        self.sides.append(self.array[9])
        self.sides.append(np.rot90(self.array)[0])
        self.sides.append(np.rot90(self.array)[9])
        self.sides.append(np.fliplr(self.array)[0])
        self.sides.append(np.fliplr(self.array)[9])
        self.sides.append(np.fliplr(np.rot90(self.array))[0])
        self.sides.append(np.fliplr(np.rot90(self.array))[9])
        #matching tiles
        self.matches = []
        self.matched_side_count = 0
        #location in picture
        #[x,y,rot,flip]
        self.loc = {'x':-1,'y':-1,'rot':0,'flip':False}
```

```python
tiles = {}
for ii in range(len(input)):
    if input[ii][:4] == 'Tile':
        tile_id = input[ii].split()[1][:-1]
        current_tile = tile(tile_id,np.asarray([list(row) for row in input[ii+1:ii+11]]))
        tiles[tile_id] = current_tile
        
```

```python
tiles['1051'].sides
```

```python
#assign matches
for tile in tiles:
    matched_sides = [0 for i in range(8)]
    for possible_match in tiles:
        if tiles[possible_match].id != tiles[tile].id:
            current_match = False
            for x, side in enumerate(tiles[tile].sides):
                for candidate in tiles[possible_match].sides:
                    if all(side == candidate):
                        current_match = True
                        matched_sides[x] = 1
            if current_match:
                tiles[tile].matches.append(tiles[possible_match].id)
    tiles[tile].matched_side_count = sum(matched_sides)
```

```python
print(tiles['1051'].matches, tiles['1051'].matched_side_count)
```

```python
[len(tiles[tile].matches) for tile in tiles]
```

```python
#corners could only have 4 matches...
corners = [tiles[tile].id for tile in tiles if tiles[tile].matched_side_count == 4]
print(corners)
```

```python
#product of corners is answer
product = 1
for j in corners:
    product *= int(j)
    
print(product)
```

```python
def find_match(tile_id,side):
    for matching_tile_id in tiles[tile_id].matches:
        for ii, candidate in enumerate(tiles[matching_tile_id].sides):
            if all(side == candidate):
                return(matching_tile_id,ii, candidate)
```

```python
#sea monster
def is_monster(array,x,y):
    monster = np.asarray([list('                  # '),list('#    ##    ##    ###'),list(' #  #  #  #  #  #   ')])
    if sum(sum(monster == array[y:y+3,x:x+20])) == 15:
        return True
    #print(np.where(array[y:y+3,x:x+20] == '#'))
```
