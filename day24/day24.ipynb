Part 1: count flipped hex tiles

```python
input=open("input.txt")
input = input.readlines()
input = [x.strip() for x in input]
print(len(input))
print(len(input[1]))
print(input[:10])
```

```python
#parse input
instructions = []
for line in input:
    instruction = []
    ii = 0
    while ii < len(line):
        if line[ii] in ('e','w'):
            instruction.append(line[ii])
            ii += 1
        else:
            instruction.append(line[ii:ii+2])
            ii += 2
    instructions.append(instruction)
    
```

```python
#use rectangular coords for mapping
# e & w will act as usual: +/-1 in x
# ne & sw will act +/-1 in y but no change in x
# nw & se will act +/-1 in y and in x
instructions
```

```python
#initiate
black_tiles = set()
x, y = 0,0
```

```python

for instruction in instructions:
    #find coordinate 
    newx, newy = x, y
    for step in instruction:
        if step == 'e':
            newx += 1
        elif step == 'w':
            newx -= 1
        elif step == 'ne':
            newy += 1
        elif step == 'se':
            newx += 1
            newy -= 1
        elif step == 'sw':
            newy -= 1
        elif step == 'nw':
            newx -= 1
            newy += 1
        else:
            print('couldn\'t parse instruction')
    print((x, y), 'to',  (newx,newy), instruction)
    if (newx, newy) in black_tiles:
        print('already seen ', (newx,newy))
        black_tiles.remove((newx,newy))
    else:
        black_tiles.add((newx,newy))
    #x,y = newx, newy
        
            
```

Part 2: 100 iterations:

Any black tile with zero or more than 2 black tiles immediately adjacent to it is flipped to white.

Any white tile with exactly 2 black tiles immediately adjacent to it is flipped to black.

```python
black_neighbours = {tile:0 for tile in black_tiles}
for tile in black_tiles:
    #add count of neigbours to black_neighbours
    x = tile[0]
    y = tile[1]
    #nw, ne, w, e, sw, se
    neighbours = [(x-1,y+1),(x,y+1),(x-1,y),(x+1,y),(x,y-1),(x+1,y-1)]
    for neighbour in neighbours:
        if neighbour in black_neighbours:
            black_neighbours[neighbour] += 1
        else:
            black_neighbours[neighbour] = 1
```

```python
for tile in black_neighbours:
    if black_neighbours[tile] == 2:
        black_tiles.add(tile)
    elif (black_neighbours[tile] == 0 or black_neighbours[tile] > 2) and tile in black_tiles:
        black_tiles.remove(tile)
```
