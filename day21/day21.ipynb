Part 1: find allergen-safe ingredients

```python
input=open("input.txt")
input = input.readlines()
input = [x.strip() for x in input]
print(len(input))
print(len(input[1]))
print(input[:10])
```

```python
ingredient_lists = []
for row in input:
    parsed = [part.split() for part in row.split(' (contains ')]
    for ii in range(len(parsed[1])):
        parsed[1][ii] = parsed[1][ii].strip(' ,)')
    ingredient_lists.append(parsed)
    
```

```python
ingredient_lists
```

```python
allergens = set()
for row in ingredient_lists:
    allergens.update(row[1])
```

```python
ingredients = set()
for row in ingredient_lists:
    ingredients.update(row[0])
```

```python
print(allergens)
print(ingredients)
```

```python
[list[1] for list in ingredient_lists]
```

```python
#counts
allergens = {'dairy':0, 'eggs':0, 'fish':0, 'peanuts':0, 'sesame':0, 'shellfish':0, 'soy':0, 'wheat':0}
for row in ingredient_lists:
    for allergen in row[1]:
        allergens[allergen] += 1
```

```python
allergens
```

```python
#find possible matches
allergen_matches = {'dairy':set(), 'eggs':set(), 'fish':set(), 'peanuts':set(), 'sesame':set(), 'shellfish':set(), 'soy':set(), 'wheat':set()}
for allergen in allergens:
    for ingredient_list in ingredient_lists:
        if allergen in ingredient_list[1]:
            if allergen_matches[allergen] == set():
                allergen_matches[allergen].update(ingredient_list[0])
            else:
                allergen_matches[allergen] = allergen_matches[allergen].intersection(ingredient_list[0])
```

```python
allergen_matches
```

```python
allergen_matches = {'dairy':'lmxt', 
                    'eggs':'rggkbpj', 
                    'fish':'mxf', 
                    'peanuts':'gpxmf', 
                    'sesame':'nmtzlj', 
                    'shellfish':'dlkxsxg', 
                    'soy':'fvqg', 
                    'wheat':'dxzq'}
```

```python
#counts
count = 0
for ingredient_list in ingredient_lists:
    for item in ingredient_list[0]:
        if item not in allergen_matches.values():
            count += 1
```
